````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

````python
dataset = pd.read_csv('50_Startups.csv')
````

````python
dataset
````

````python

x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,4:5].values
````

````python
x
````

````python
y
````

````python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
from sklearn.compose import ColumnTransformer
````

````python
transformer = ColumnTransformer(
transformers = [
    (
    "sda",
        OneHotEncoder(),
        [3]
    ),
],
    remainder = 'passthrough'
)
````

````python
x=transformer.fit_transform(x)
````

````python
x
````

````python
x = x[:,1:]
````

````python
x
````

## Building the optimal model using Backward Elimination

# multiple linear regression

 ```
 y = B0 + B1x1 + B2x2 + B3x3 + ....
 
 ```
 
 # In use of backward Elimination 
 
 ```
 y = B0x0 + B1x1 + B2x2 + B3x3 + ....
 ```
 
 here `x0` is a set of `1` array where every `x = 1`

````python
x = np.append(arr=np.ones((50,1)).astype(int),values = x,axis = 1)
````

````python
x
````

````python
import statsmodels.api as sm
````

## At first take all independent elements

````python
x_opt = np.array(x[:, [0, 1, 2, 3, 4, 5]], dtype=float)
````

## Fit model with them

````python
regressor_ols = sm.OLS(endog=y,exog=x_opt).fit()
````

## Lets see the summary to find the indipendent variable which has p-value more then the significant level (sl = 0.05 or 5% lets assume)

````python
regressor_ols.summary()
````

## X2 has the height p-value  more then sl(0.05) so remove it

````python
x_opt = np.array(x[:,[0,1,3,4,5]],dtype=float)
regressor_ols = sm.OLS(endog=y,exog=x_opt).fit()
regressor_ols.summary()
````

````python
x
````

## x1 has the height p-value more then sl(0.05) so we will remove it

````python
x_opt = np.array(x[:,[0,3,4,5]],dtype=float)
regressor_ols = sm.OLS(endog=y,exog=x_opt).fit()
regressor_ols.summary()
````

````python
x
````

## X1 has  height p-value   more then sl(0.05) 

````python
x_opt = np.array(x[:,[0,3,5]],dtype=float)
regressor_ols = sm.OLS(endog=y,exog=x_opt).fit()
regressor_ols.summary()
````

````python
x
````

## X1 has  height p-value   more then sl(0.05) 

````python
x_opt = np.array(x[:,[0,3]],dtype=float)
regressor_ols = sm.OLS(endog=y,exog=x_opt).fit()
regressor_ols.summary()
````

## As we can see in the above output image, only two variables are left. So only the R&D independent variable is a significant variable for the prediction. So we can now predict efficiently using this variable.

````python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test = tts(x[:,[3]],y,test_size = 0.25,random_state=5)
````

````python
x_train
````

````python
x_test
````

````python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
y_pred=regressor.predict(x_test)
````

````python
y_pred
````

````python
y_test
````

````python
print('Train Score: ', regressor.score(x_train, y_train))  
print('Test Score: ', regressor.score(x_test, y_test))
````
